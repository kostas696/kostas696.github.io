# 1. Introduction

### Background Information

The **Medical Cost Personal Datasets** is a public dataset used in the book "Machine Learning with R" by Brett Lantz. The dataset has been cleaned and recoded to match the format in the book. It contains information about medical insurance charges for individuals based on various factors. The original dataset is available on Kaggle [Medical Cost Personal Datasets](https://www.kaggle.com/datasets/mirichoi0218/insurance).



### Objective of the Analysis

The objective of this analysis is to predict medical insurance charges for individuals using the provided dataset. By exploring the relationships between variables such as **age, gender, BMI, number of children, smoking status, residential region, and medical costs**, we aim to develop a predictive model that can accurately estimate insurance charges. This analysis will employ various regression algorithms and evaluate their performance to identify the most effective model for predicting insurance costs.

By accomplishing this objective, we can gain insights into the factors that significantly impact medical insurance charges and potentially contribute to more accurate cost estimations in the insurance industry.

**Inspiration**:
The main inspiration behind this project is to leverage machine learning techniques to accurately predict insurance costs. By exploring the relationships between various factors and individual medical charges, we aim to build a model that can provide valuable predictions for insurance companies and individuals alike. The outcomes of this analysis can have practical applications in cost estimation, risk assessment, and informed decision-making in the insurance domain.

Throughout this project, we will follow a structured approach, including data exploration and preprocessing, univariate, bivariate, and multivariate analyses, as well as implementing and evaluating multiple regression algorithms. The goal is to develop a robust predictive model that accurately estimates medical insurance charges based on the available dataset.

### Dataset Information

***Age:*** Age is an essential factor in determining health insurance charges. As individuals grow older, they tend to be at a higher risk of health issues, which can lead to increased medical expenses. Consequently, insurance charges often rise with age.

***Sex:*** Gender can also play a role in health insurance costs. Women may have higher insurance charges due to factors such as pregnancy, childbirth, and women's specific healthcare needs. However, the impact of gender on insurance charges can vary based on other factors within the dataset.

***BMI (Body Mass Index)***: BMI is a measure that provides insight into an individual's body weight relative to their height. Higher BMI values are often associated with increased health risks and potential healthcare expenses. As a result, individuals with higher BMI values may experience higher insurance charges.

***Children:*** The number of children covered by health insurance or the number of dependents can influence insurance charges. Insuring more individuals within a family plan generally leads to higher costs.

***Smoker:*** Smoking is a well-known risk factor for various health conditions, including respiratory diseases, heart disease, and cancer. As a result, smokers typically face higher insurance charges due to the increased likelihood of healthcare utilization and associated expenses.

***Region:*** The residential area of the insurance beneficiary can impact insurance charges. Different regions may have varying healthcare costs, cost of living, and prevalence of certain health conditions, all of which can contribute to differences in insurance charges.

***Charges:*** The "charges" column represents the individual medical costs billed by health insurance. It serves as the dependent variable in this dataset. The goal of the analysis is to understand how the independent variables (age, sex, BMI, children, smoker, region) relate to these charges and build a predictive model for estimating insurance costs accurately.

By considering these factors and analyzing their relationships with insurance charges, we can gain valuable insights into the determinants of healthcare expenses. This understanding can aid in predicting insurance costs more effectively and making informed decisions in the insurance industry.

# 5. Bivariate Analysis

### 5.1 Age vs. Charges

In this section we could search a bit more our dataset to gain more insights. For example we could group numerical values of a column such as 'age' into age brackets (e.g., 20s, 30s, 40s, etc.) and analyze the distribution of charges within each age group.

```python
# Grouping 'age' into age brackets
age_bins = [18, 30, 40, 50, 60, 70]
age_labels = ['20s', '30s', '40s', '50s', '60s']

df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Analyzing the distribution of charges within each age group
sns.boxplot(x=df['age_group'], y=df['charges'])
plt.xlabel('Age Group')
plt.ylabel('Charges')
plt.title('Distribution of Charges by Age Group')
plt.show()
```

```python
df['age_group'].value_counts()
```

The majority of the beneficiaries fall into the '20s' age group, followed by the '40s', '50s', '30s', and '60s' age groups.
This distribution suggests that the dataset contains a relatively higher number of younger individuals.

The box plot indicates differences in the distribution of charges between the age groups. Older age groups tend to be associated with higher charges, and specifically, the 50s age group shows higher charges than expected compared to the other age groups(many outliers). In this case, the higher charges observed outside the whiskers for the 50s age group indicate that there is a subset of individuals within this age group who experience particularly high medical expenses.

There could be several reasons why the 50s age group shows higher charges compared to the other age groups, such us increased health risks, higher incidence of chronic conditions, often healthcare utilization etc. However, it's important to note that further investigation is required to understand the specific factors contributing to these higher charges and to determine if they are truly outliers or reflect a distinct characteristic of the 50s age group.

To further support and quantify these observations, we can use statistical tests such as **ANOVA (Analysis of Variance)**. ANOVA allows us to assess if there are statistically significant differences in charges among different age groups determining if the observed variation in charges across age groups is statistically significant or if it could be due to random variation. In this case, the ANOVA results confirmed that there are significant differences in charges among the age groups.

```python
# Perform ANOVA for charges by age group
model = ols('charges ~ age_group', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

The ANOVA test examines whether there are statistically significant differences in charges among the different age groups. The test output includes the sum of squares (sum_sq), degrees of freedom (df), F-statistic (F), and the associated p-value (PR(>F)).

Sum of Squares: The sum of squares measures the total variability in charges that can be explained by the age group variable. In this case, the sum of squares for the age_group variable is 4.0.

Degrees of Freedom: The degrees of freedom indicate the number of independent pieces of information available for estimating variability. For the age_group variable, there are 4 degrees of freedom (df) because there are 5 age groups.

F-Statistic: The F-statistic is a ratio of the between-group variability (explained variability) to the within-group variability (unexplained variability). In this case, the F-statistic is 120.56.

p-value: The p-value associated with the F-statistic represents the probability of obtaining results as extreme as the observed results, assuming that there are no true differences among the age groups. A low p-value indicates strong evidence against the null hypothesis (no differences) and suggests that the observed differences are statistically significant. In this case, the p-value is extremely small (9.16e-88), indicating highly significant differences among the age groups.

Therefore, based on the ANOVA results, we can confidently conclude that there are statistically significant differences in charges among the age groups. This finding supports the initial observation of outliers in the 50s age group and provides further evidence that age plays a role in determining the charges in the dataset.

### 5.2 Sex vs. Charges

Here, with the help of the plot below we have a visual representation of the distribution of charges for each gender, highlighting any potential differences. This plot allows us to visually assess any differences in healthcare costs between the two genders.

```python
# Create a box plot to compare charges between males and females
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='sex', y='charges')
plt.xlabel('Sex')
plt.ylabel('Charges')
plt.title('Charges by Sex')
plt.show()
```

Based on the boxplot, it appears that the box representing charges for males is larger than the box representing charges for females. This suggests that the interquartile range (IQR) of charges for males is larger, indicating more variability in charges among males.

Next, we perform an independent t-test to determine if there is a statistically significant difference in charges between males and females.

```python
# Perform t-test to compare charges between males and females
charges_male = df[df['sex'] == 'male']['charges']
charges_female = df[df['sex'] == 'female']['charges']
t_statistic, p_value = stats.ttest_ind(charges_male, charges_female)

# Print the t-statistic and p-value
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

The independent t-test is used to compare the means of two groups and assess whether the observed difference is statistically significant. In this case, the t-test is conducted to determine if there is a significant difference in charges between males and females.

T-Statistic: The t-statistic measures the difference between the means of the two groups relative to the variation within each group. The t-statistic in this case is 0.413.

P-Value: The p-value associated with the t-statistic indicates the probability of obtaining the observed results (or more extreme) if there were no true difference between the means of the two groups. A higher p-value suggests weak evidence against the null hypothesis (no difference). In this case, the p-value is 0.679.

Based on the provided results, the t-statistic is close to zero, and the p-value is relatively high (0.679). This indicates that there is insufficient evidence to conclude a statistically significant difference in charges between males and females. Therefore, we fail to reject the null hypothesis, suggesting that there is no significant difference in healthcare costs between the two genders in the dataset.

It's important to note that if there are other relevant factors or variables that could influence charges, it might be necessary to consider them in the analysis or perform further investigations to gain a comprehensive understanding of the relationship between gender and charges.

These observations can be further investigated and validated through additional statistical analysis. The **independent t-test** assesses if there is a significant difference in charges between the two groups. P-Value: The p-value associated with the t-statistic indicates the probability of obtaining the observed results (or more extreme) if there were no true difference between the means of the two groups. A lower p-value suggests strong evidence against the null hypothesis (no difference). In this case, the p-value is 0.0003.

Based on the provided results, the t-statistic is positive (3.62) and the p-value is relatively small (0.0003). This indicates that there is strong evidence to suggest a statistically significant difference in charges between the 'Obese' and 'Normal Weight' BMI categories. Therefore, we reject the null hypothesis and conclude that there are significant differences in healthcare costs between these two BMI categories.

These findings align with the boxplot that visualized the distribution of charges across BMI categories. The boxplot showed potential differences in charges between the BMI categories, with the 'Obese' category having a larger interquartile range (IQR) compared to the 'Normal Weight' category.

These results suggest that BMI category is associated with differences in healthcare charges. Additional hypothesis tests can be conducted to compare charges between other BMI categories and explore further relationships within the dataset.

The box plot provides a visual representation of the distribution of charges for each number of children, allowing us to observe any variations in healthcare costs. The ANOVA test helps quantify the statistical significance of these differences.

```python
# Perform ANOVA test to assess if there are significant differences in charges among different number of children
groups = df.groupby('children')['charges'].apply(list)
f_statistic, p_value = stats.f_oneway(*groups)

# Print the F-statistic and p-value
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)
```

F-Statistic: The F-statistic measures the ratio of between-group variability to within-group variability. In this case, the F-statistic is 9.16.

P-Value: The p-value associated with the F-statistic indicates the probability of obtaining the observed results (or more extreme) if there were no true differences among the groups. A lower p-value suggests strong evidence against the null hypothesis (no differences). In this case, the p-value is 1.39e-08, which is extremely small.

Based on the provided results, the F-statistic is positive (9.16) and the p-value is significantly small (1.39e-08). This indicates strong evidence to suggest statistically significant differences in charges among different numbers of children. Therefore, we reject the null hypothesis and conclude that there are significant variations in healthcare costs based on the number of children.

These findings align with the box plot that visualized the distribution of charges across different numbers of children. The varying sizes of the boxes and the differences in median lines indicate potential differences in charges among the groups.

```python
# Perform t-test to compare charges between smokers and non-smokers
charges_smoker = df[df['smoker'] == 'yes']['charges']
charges_non_smoker = df[df['smoker'] == 'no']['charges']
t_statistic, p_value = stats.ttest_ind(charges_smoker, charges_non_smoker)

# Print the t-statistic and p-value
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

T-Statistic: The t-statistic measures the difference between the means of the smoker and non-smoker groups relative to the variation within each group. In this case, the t-statistic is 32.319.

P-Value: The p-value associated with the t-statistic indicates the probability of obtaining the observed results (or more extreme) if there were no true difference between the means of the two groups. A lower p-value suggests strong evidence against the null hypothesis (no difference). In this case, the p-value is 1.74e-169, which is extremely small.

Based on the provided results, the t-statistic is significantly large (32.319), and the p-value is extremely small (1.74e-169). This indicates strong evidence to suggest a statistically significant difference in charges between smokers and non-smokers. Therefore, we reject the null hypothesis and conclude that there are significant differences in healthcare costs between these two groups.

In summary, the analysis suggests that smokers tend to have significantly higher healthcare charges compared to non-smokers. This finding underscores the impact of smoking on healthcare expenses and highlights the importance of addressing smoking-related health risks.

### 5.6 Region vs. Charges

The following box plot provides a visual representation of the distribution of charges for each region, allowing us to observe any variations in healthcare costs across different geographic areas.

```python
# Create a box plot to compare charges across different regions
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='region', y='charges')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.title('Charges by Region')
plt.show()
```

 It appears that there are some differences in charges across different regions. The box representing charges for the southeast region is the largest, followed by the northeast region.

Next, we perform a one-way analysis of variance (ANOVA) to determine if there are significant differences in charges among the different regions.

```python
# Perform one-way ANOVA to assess if there are significant differences in charges among different regions
groups = [df[df['region'] == 'southeast']['charges'],
          df[df['region'] == 'southwest']['charges'],
          df[df['region'] == 'northeast']['charges'],
          df[df['region'] == 'northwest']['charges']]
f_statistic, p_value = stats.f_oneway(*groups)

# Print the F-statistic and p-value
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)
```

F-Statistic: The F-statistic measures the ratio of between-group variability to within-group variability. In this case, the F-statistic is 1.464.

P-Value: The p-value associated with the F-statistic indicates the probability of obtaining the observed results (or more extreme) if there were no true differences among the groups. A higher p-value suggests weak evidence against the null hypothesis (no differences). In this case, the p-value is 0.223.

Based on the provided results, the F-statistic is relatively small (1.464), and the p-value is relatively high (0.2225). This suggests that there is not enough evidence to conclude statistically significant differences in charges among the different regions. Therefore, we fail to reject the null hypothesis and conclude that there are no significant variations in healthcare costs based on region.

These findings align with the box plot that visualized the distribution of charges across different regions. While there are some differences in the size of the boxes, indicating potential variations in charges, the ANOVA test results indicate that these differences are not statistically significant.

### 8.3 Hyperparameter Tuning

By performing hyperparameter tuning, we can fine-tune the models and achieve better results, potentially improving their accuracy, reducing overfitting, and enhancing their ability to generalize to new data. We will perform hyperparameter tuning for each model in the pipeline using GridSearchCV.
GridSearchCV is applied on the model using the specified hyperparameter grid, 5-fold cross-validation, and 'neg_mean_squared_error' scoring. With the optimized hyperparameters, we then evaluate their performance using MSE and R-squared scores.

```python
# Define the hyperparameter grids for each model
linear_regression_params = {}
ridge_regression_params = {'model__alpha': [0.1, 1.0, 10.0]}
lasso_regression_params = {'model__alpha': [0.1, 1.0, 10.0]}
elasticnet_regression_params = {'model__alpha': [0.1, 1.0, 10.0], 'model__l1_ratio': [0.1, 0.3, 0.5]}
polynomial_regression_params = {'poly__degree': [2, 3, 4]}
svr_params = {'model__C': [0.1, 1.0, 10.0], 'model__kernel': ['linear', 'rbf']}
random_forest_params = {'model__n_estimators': [100, 200, 300]}
gradient_boosting_params = {'model__n_estimators': [100, 200, 300], 'model__learning_rate': [0.1, 0.01, 0.001]}

# Perform hyperparameter tuning for each model
models = [
    ('Linear Regression', linear_regression_pipeline, linear_regression_params),
    ('Ridge Regression', ridge_regression_pipeline, ridge_regression_params),
    ('Lasso Regression', lasso_regression_pipeline, lasso_regression_params),
    ('ElasticNet Regression', elastic_net_pipeline, elasticnet_regression_params),
    ('Polynomial Regression', polynomial_pipeline, polynomial_regression_params),
    ('Support Vector Regression', svr_pipeline, svr_params),
    ('Random Forest Regression', random_forest_pipeline, random_forest_params),
    ('Gradient Boosting Regressor', gradient_boosting_pipeline, gradient_boosting_params)
]

for model_name, model, param_grid in models:
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)

    best_params = grid_search.best_params_
    model.set_params(**best_params)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(model_name)
    print("Best Hyperparameters:", best_params)
    print("Mean Squared Error:", mse)
    print("R^2 Score:", r2)
    print("--------------------")
```

Compared to the previous results, we can see major improvements in the performance of Lasso Regression, and ElasticNet Regression after hyperparameter tuning. These models achieved better Mean Squared Error and R-squared scores, indicating improved accuracy and predictive power.

Among all the models, the Gradient Boosting Regressor performed the best after hyperparameter tuning. It achieved the lowest Mean Squared Error and the highest R-squared score, indicating superior performance in terms of prediction accuracy.
